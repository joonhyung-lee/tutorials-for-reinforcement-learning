```python
import random
import gym
import numpy as np
from IPython.display import clear_output
from time import sleep

import torch
from torch.distributions import Categorical
from torch.nn import Module, Linear
import torch.nn.functional as F



class PolicyNetwork(Module):
    def __init__(self):
        super().__init__()
        self.fc = Linear(500, 6)
    def forward(self, x):
        x = self.fc(x)
        return torch.nn.functional.softmax(x, dim=-1)
    
    

# network and optimizer

pi = PolicyNetwork()
Actor_optimizer = torch.optim.Adam(pi.parameters(), lr=0.0001)

def onehot(s):
    encoded = np.zeros(500)
    encoded[s] = 1
    return encoded
        
def update_pi():
    loss = 0
    for t in range(len(self.states)):
        state = self.states[t]
        action = self.actions[t]
        loss = loss + (gamma**t) * Q[state,action] * pi(state)[action].log()
    Actor_optimizer.zero_grad()
    loss.backward()
    Actor_optimizer.step()       
    

# env = gym.make('FrozenLake-v0')
env = gym.make("Taxi-v3")
score = 0
episode = 0
state_size = 500
action_size = 6
MAX_EPISODES = 100000
Q = np.zeros([state_size, action_size])
states = []
actions = []
gamma = 0.9
alpha = 0.0001

state = env.reset()
probs = pi(onehot(state))
action = torch.multinomial(probs, 1).item()

while episode < MAX_EPISODES:  # episode loop
    #env.render()
    next_state, reward, done, info = env.step(action)  # take a random action
    score = score + reward
    states.append(state)
    actions.append(action)
    probs = pi(onehot(next_state))
    next_action = torch.multinomial(probs, 1).item()
    
    if done:
        Q[state,action] = (1 - alpha) * Q[state,action] + alpha * (reward)
    else:
        Q[state,action] = (1 - alpha) * Q[state,action] + alpha * (reward + gamma * Q[next_state,next_action])
    
    state = next_state
    action = next_action
    


    if done:
        episode = episode + 1
        update_pi()
        states = []
        actions = []
        if episode % 100 == 0:
            clear_output(wait=True)
            print('Episode: {} Score: {}'.format(episode, score))
        score = 0
        state = env.reset()
        
# TEST     
episode = 0
state = env.reset()     
MAX_EPISODES = 10
while episode < MAX_EPISODES:  # episode loop
    env.render()
    probs = pi(onehot(state))
    action = torch.multinomial(probs, 1).item()
    next_state, reward, done, info = env.step(action)  # take a random action
    score = score + reward
    state = next_state
    sleep(0.01)
    

    if done:
        episode = episode + 1
        if episode % 100 == 0:
            clear_output(wait=True)
            print('Episode: {} Score: {}'.format(episode, score))
        score = 0
        state = env.reset()
        
```
