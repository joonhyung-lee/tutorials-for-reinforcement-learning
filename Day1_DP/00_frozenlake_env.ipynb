```python
import gym
from time import sleep

env = gym.make('FrozenLake-v1')

print("Action Space {}".format(env.action_space))
print(f"Dimension of Action space: {env.action_space.n}")
print("State Space {}".format(env.observation_space))
print(f"Dimension of State space: {env.observation_space.n}")
env = gym.wrappers.TimeLimit(env, max_episode_steps = 5)    # 5 steps per episode
env.reset()

s=1
a=1
# Probability of transition to next state
print(env.P[s][a])

done = False
env.render()
while not done:
    new_state, reward, done, info = env.step(env.action_space.sample())
    env.render()
    print(reward, new_state)
    sleep(.1)
```
