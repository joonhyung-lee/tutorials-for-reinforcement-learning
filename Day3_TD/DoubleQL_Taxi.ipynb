```python
import gym
import numpy as np
from time import sleep
import random
from IPython.display import clear_output
# env = gym.make('FrozenLake-v0')
env = gym.make("Taxi-v3")

Qa = np.zeros([env.observation_space.n, env.action_space.n])
Qb = np.zeros([env.observation_space.n, env.action_space.n])
alpha = 0.1
gamma = 1
epsilon = 0.1

for i in range(1, 100001):
    state = env.reset()
    reward= 0
    done = False
    while not done:
        if random.uniform(0, 1) < epsilon:
            action = env.action_space.sample() # Explore action space
        else:
            action = np.argmax(Qa[state]+Qb[state]) # Exploit learned values
        next_state, reward, done, info = env.step(action)
        if done:
            if random.uniform(0, 1) < 0.5:
                Qa[state, action] = (1 - alpha) * Qa[state, action] + alpha * (reward)
            else:
                Qb[state, action] = (1 - alpha) * Qb[state, action] + alpha * (reward)
        else:
            if random.uniform(0, 1) < 0.5:
                Qa[state, action] = (1 - alpha) * Qa[state, action] + alpha * (reward + gamma * Qb[next_state,np.argmax(Qa[next_state])])
            else:
                Qb[state, action] = (1 - alpha) * Qb[state, action] + alpha * (reward + gamma * Qa[next_state,np.argmax(Qb[next_state])])
        state = next_state
    if i % 100 == 0:
        clear_output(wait=True)
        print('Episode: {0}'.format(i))
print("Training finished.\n")

"""Evaluate agent's performance after Q-learning"""
episodes = 100
for _ in range(episodes):
    state = env.reset()
    done = False
    env.render()
    while not done:
        action = np.argmax(Qa[state])
        state, reward, done, info = env.step(action)
        env.render()

```
