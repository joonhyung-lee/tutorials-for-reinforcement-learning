```python
import random
import gym
import numpy as np

env = gym.make('LunarLander-v2')

total_score_memory = []
out_loop = 0
while out_loop < 15:  # episode loop
    # TEST     
    episode = 0
    total_score = 0

    state = env.reset()     
    while episode < 3000:  # episode loop
        action = env.action_space.sample()
        next_state, reward, done, info = env.step(action)  # take a random action
        state = next_state
        total_score = total_score + reward

        if done:
            episode = episode + 1
            if episode % 100 == 0:
                print('Episode: {} Total score: {}'.format(episode, total_score))
            state = env.reset()
    env.close()   
    
    total_score_memory.append(total_score)
    out_loop = out_loop +1

avg = sum(total_score) / len(total_score)
print(avg)
```
